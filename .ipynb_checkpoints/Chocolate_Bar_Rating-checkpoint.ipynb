```python
import pandas as pd
import numpy as np
#import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt

path = "./data/flavors_of_cacao.csv"
df = pd.read_csv(path)
```

```python
import re
## Text preparation (correction) func
def txt_prep(text):
    replacements = [
        ['-', ', '], ['/ ', ', '], ['/', ', '], ['\(', ', '], [' and', ', '], [' &', ', '], ['\)', ''],
        ['Dom Rep|DR|Domin Rep|Dominican Rep,|Domincan Republic', 'Dominican Republic'],
        ['Mad,|Mad$', 'Madagascar, '],
        ['PNG', 'Papua New Guinea, '],
        ['Guat,|Guat$', 'Guatemala, '],
        ['Ven,|Ven$|Venez,|Venez$', 'Venezuela, '],
        ['Ecu,|Ecu$|Ecuad,|Ecuad$', 'Ecuador, '],
        ['Nic,|Nic$', 'Nicaragua, '],
        ['Cost Rica', 'Costa Rica'],
        ['Mex,|Mex$', 'Mexico, '],
        ['Jam,|Jam$', 'Jamaica, '],
        ['Haw,|Haw$', 'Hawaii, '],
        ['Gre,|Gre$', 'Grenada, '],
        ['Tri,|Tri$', 'Trinidad, '],
        ['C Am', 'Central America'],
        ['S America', 'South America'],
        [', $', ''], [',  ', ', '], [', ,', ', '], ['\xa0', 'FILL'],[',\s+', ','],
        [' Bali', ',Bali']
    ]
    for i, j in replacements:
        text = re.sub(i, j, text)
    return text
```

```python
# Replace country feature
df['Broad Bean\nOrigin'] = df['Broad Bean\nOrigin'].str.replace('.', '').apply(txt_prep)
```

```python
# define blend feature
#１のnp.logical_or: 空白の国か、バーの種類にstr.contains(',|(blend)|;')ならTrue
#２のnp.logical_or: ","が含まれてる国を"blend"に追加
#np.where: Trueなら１、Falseなら0
df['is_blend'] = np.where(np.logical_or(np.logical_or(df['Specific Bean Origin\nor Bar Name'].str.lower().str.contains(',|(blend)|;'),
                                                      df['Broad Bean\nOrigin']=='FILL'),df['Broad Bean\nOrigin'].str.lower().str.contains(',')), 1, 0)
## How many blends/pure cocoa?
df['is_blend'].value_counts()
```

```python
# We need to make some replacements
df['Company\nLocation'] = df['Company\nLocation']\
.str.replace('Amsterdam', 'Holland')\
.str.replace('U.K.', 'England')\
.str.replace('Niacragua', 'Nicaragua')\
.str.replace('Domincan Republic', 'Dominican Republic')

df['Company\nLocation'].sort_values().unique()
```

```python
# Define domestic feature
df['is_domestic'] = np.where(df['Broad Bean\nOrigin'] == df['Company\nLocation'], 1, 0)
df['is_domestic'].value_counts()
```

```python
df['Bean\nType'] = df['Bean\nType'].fillna('Blend')
```

```python
import re
from tokenize import Tokenizer
from nltk.stem.lancaster import LancasterStemmer
stemmer=LancasterStemmer()

def stem(l):
    #tokenizer = RegexpTokenizer(r'\w+')
    token_words = Tokenizer.tokenize(l)
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(stemmer.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)

df["stemmed_bean_origin"] = df.dropna().country.apply(lambda l: stem(l))
```

```python
def objective(trial):
    # トレーニングデータとテストデータを分割
    # パラメータの指定
    params = {
        'iterations' : trial.suggest_int('iterations', 50, 300),                         
        'depth' : trial.suggest_int('depth', 3, 10),                                       
        'learning_rate' : trial.suggest_loguniform('learning_rate', 0.1, 1),               
        'random_strength' :trial.suggest_int('random_strength', 0, 100),                       
        'bagging_temperature' :trial.suggest_loguniform('bagging_temperature', 0.01, 100.00), 
        'od_type': trial.suggest_categorical('od_type', ['IncToDec', 'Iter']),
        'od_wait' :trial.suggest_int('od_wait', 10, 50)
    }

    # 学習
    model = CatBoostRegressor(**params,loss_function='RMSE')
    model.fit(train_pool)
    # 予測
    preds = model.predict(test_pool)
    pred_labels = np.rint(preds)
    # 精度の計算
    #accuracy = sklearn.metrics.accuracy_score(y_test, pred_labels)
    #return 1.0 - accuracy
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    return rmse
```

```python
def model(param_cocoa_p, param_country):
    pam = pd.DataFrame([{'Cocoa\nPercent' : float(param_cocoa_p), 'Broad Bean\nOrigin' : str(param_country)}])
    df2 = pd.concat([df, pam]).reset_index(drop=True)
    
    mean = np.nanmean(df2['Cocoa\nPercent'], axis=0)
    std = np.nanstd(df2['Cocoa\nPercent'], axis=0)
    df2['Cocoa\nPercent'] = (df2['Cocoa\nPercent'] - mean)/std
    
    df2['Broad Bean\nOrigin'] = pd.get_dummies(df2['Broad Bean\nOrigin'], drop_first=True)
    
    #model_input = df2[['Cocoa\nPercent', 'Broad Bean\nOrigin']]
    #X = model_input.dropna(how='any', axis=1) 
    X = df2[['Cocoa\nPercent', 'Broad Bean\nOrigin']]
    y = df2['Style_id']
    df_len = df2.shape[0]
    X_train = X[:df_len]
    X_test = X[df_len:]
    y_train = y[:df_len]
    y_test = y[df_len:]
    #X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
    
    #from sklearn.neighbors import KNeighborsClassifier
    from sklearn.ensemble import RandomForestRegressor

    #best_params = {'n_neighbors': 30}
    #best_n = best_params['n_neighbors']
    #neigh = KNeighborsClassifier(n_neighbors=best_n)
    random = RandomForestRegressor(n_estimators=20, random_state=0)
    model = random.fit(X_train, y_train)
    pred_id = model.predict(X_test)
    return pred_id
```
